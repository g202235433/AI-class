```python
from sklearn.datasets import load_iris

iris = load_iris()
iris.keys()
```

```python
iris_data = iris.data
iris_target = iris.target
```

```python
print(iris['feature_names'])
print(iris['target_names'])
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
labels = pd.DataFrame(iris.target, columns = ['speice'])
print(df)
```

```python
df.isnull().sum()
```

```python
data = pd.concat([df, labels], axis=1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(data,
             vars=['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'],
            hue ='total')
plt.show()
```

```python
sns.countplot(data = data, x="total")
plt.xlabel("total")
plt.ylabel("cm")
plt.title("total")
```
