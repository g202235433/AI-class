```python
from sklearn.datasets import load_breast_cancer

import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import StandardScaler, LabelEncoder

from sklearn.neighbors import KNeighborsClassifier

breast_cancer_data = load_breast_cancer()
```

```python
df_features = breast_cancer_data.feature_names
```

```python
df_features
```

```python
df_data = pd.DataFrame(breast_cancer_data.data, columns = df_features)

df_labels = pd.DataFrame(breast_cancer_data.target, columns=['label'])   
```

```python
df_data
```

```python
print(df_data[:10])

print(df_labels[:10])
```

```python
print(breast_cancer_data.target_names)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_data,df_labels, test_size = 0.2, random_state=3)
```

```python
print("X_train.shape :", X_train.shape)

print("y_train.shape :", y_train.shape)

print("X_test.shape:", X_test.shape)

print("y_test.shape:", y_train.shape)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
mymodel = KNeighborsClassifier(n_neighbors=3)
mymodel.fit(X_train, y_train)
prediction = mymodel.predict(X_test)
mymodel.score(X_test, y_test)
```

```python
from sklearn import metrics
import numpy as np

K = 10
mean_acc = np.zeros((K -1))
std_acc = np.zeros((K -1))

for n in range(1, K):
    clf = KNeighborsClassifier(n_neighbors=n).fit(X_train, y_train)
    yhat = clf.predict(X_test)
    mean_acc[n-1] = metrics.accuracy_score(y_test, yhat)
    
print(mean_acc.max())
```
